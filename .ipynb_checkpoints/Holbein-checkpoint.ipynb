```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
from skimage import io
from skimage import transform
```

```python
img = io.imread('holbein.jpg')
```

```python
plt.figure(figsize=(20,20))
plt.imshow(img)
```

```python
subimg = img[600:1000,200:800]
```

```python
plt.imshow(subimg)
```

```python
mywarp = transform.AffineTransform(shear=1.15, scale=(0.5,1))

warped = transform.warp(subimg,mywarp, output_shape=(1000,3000))

plt.figure(figsize=(15,15))
plt.imshow(warped)
```
